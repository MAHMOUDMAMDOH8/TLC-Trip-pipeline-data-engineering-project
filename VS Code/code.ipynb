```python
import io
import pandas as pd
import requests
import pyarrow.parquet as pq
import os
```

```python
urls = ['https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-01.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-07.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-02.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-08.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-03.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-09.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-04.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-10.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-05.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-11.parquet',
            'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2023-06.parquet'



    ]

Data2023 = []

for U in urls:
    response = requests.get(U)
    if response.status_code == 200:
        try:
            table = pq.read_table(io.BytesIO(response.content))
            df = table.to_pandas()
            Data2023.append(df)
        except Exception as e:
            print(f'Failed to read data from {U}: {e}')
    else:
        print(f'Failed to fetch data from {U}')

Data2023
```

```python
folder_path  = "D:\Shair storae\Project\DE Zoom camp\DE\week3\Data"
parquet_files = [f for f in os.listdir(folder_path) if f.endswith('.parquet')]
Data2022 = []
for file in parquet_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_parquet(file_path)
    Data2022.append(df)

Data2022
```

```python
combined_df = pd.concat([pd.concat(Data2022), pd.concat(Data2023)], ignore_index=True)
```

```python
combined_df.to_csv(r'D:\Shair storae\Project\DE\Uber\Vcode\Volume\All data\G_taxi.csv')
```

```python
combined_df['lpep_pickup_datetime'] = pd.to_datetime(combined_df['lpep_pickup_datetime'])
combined_df['lpep_dropoff_datetime'] = pd.to_datetime(combined_df['lpep_dropoff_datetime'])
combined_df = combined_df[(combined_df['passenger_count'] > 0) & (combined_df['trip_distance'] > 0)]
```

```python

DimDate = combined_df[['lpep_pickup_datetime','lpep_dropoff_datetime']]
DimDate['Date_key'] = DimDate.drop_duplicates
DimDate['Pick_hour'] = DimDate['lpep_pickup_datetime'].dt.hour
DimDate['Pick_day'] = DimDate['lpep_pickup_datetime'].dt.day
DimDate['Pick_month'] = DimDate['lpep_pickup_datetime'].dt.month
DimDate['Pick_year'] = DimDate['lpep_pickup_datetime'].dt.year
DimDate['Pick_weekday'] = DimDate['lpep_pickup_datetime'].dt.weekday
DimDate['drop_hour'] = DimDate['lpep_dropoff_datetime'].dt.hour
DimDate['drop_day'] = DimDate['lpep_dropoff_datetime'].dt.day
DimDate['drop_month'] = DimDate['lpep_dropoff_datetime'].dt.month
DimDate['drop_year'] = DimDate['lpep_dropoff_datetime'].dt.year
DimDate['drop_weekday'] = DimDate['lpep_dropoff_datetime'].dt.weekday
DimDate['Date_key'] = DimDate.index
```

```python
# Mapping dictionary for RatecodeID values
rate_mapping = {
    1: 'Standard rate',
    2: 'JFK',
    3: 'Newark',
    4: 'Nassau or Westchester',
    5: 'Negotiated fare',
    6: 'Group ride'
}


rate_df = pd.DataFrame(rate_mapping.items(), columns=['RatecodeID', 'Rate'])

rate_table = pd.merge(rate_df, combined_df[['RatecodeID']], on='RatecodeID', how='inner').drop_duplicates()

```

```python
Payment_mapping = {
    1: 'Credit card',
    2: 'Cash',
    3: 'No charge',
    4: 'Dispute',
    5: 'Unknown',
    6: 'Voided trip'
}
payment_df = pd.DataFrame.from_dict(Payment_mapping, orient='index', columns=['Payment_type'])
payment_df.index.name = 'Payment_ID'

# Reset index to make Payment_ID a column
payment_df.reset_index(inplace=True)
```

```python
Trip_type_mapping = {
    1: 'Street-hail',
    2: 'Dispatch'
}
Trip_type_df = pd.DataFrame.from_dict(Trip_type_mapping, orient='index', columns=['type'])
Trip_type_df.index.name = 'Trip_type_ID'
Trip_type_df.reset_index(inplace=True)


```

```python
combined_df = combined_df.rename(columns={'VendorID':'Vendor_ID','store_and_fwd_flag':'Store_Forward_fwd__flag','RatecodeID':'RateCode_ID',
                                          'PULocationID':'PuLocation_ID','DOLocationID':'DoLocation_ID','RatecodeID':'RateCode_ID',
                                          'payment_type':'Payment_ID','trip_type':'Trip_type_id'})
combined_df
```
